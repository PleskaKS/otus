# I. Numpy

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr_np = np.zeros(10)
arr_np[4] = 1
arr_np = arr_np.reshape(5,2)
arr_np
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr_np = np.arange(start=10, stop=50)
arr_np = arr_np[::-1]
arr_np % 2 == 0  
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr_np = np.arange(start=0, stop=9)
arr_np = arr_np.reshape(3,3)
arr_np
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr_np = np.random.random(24)
arr_np =arr_np.reshape(4,3,2)
arr_np, arr_np.min(), arr_np.max()
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr_1 = np.arange(start=1, stop=25)
arr_1 = arr_1.reshape(6,4)
arr_2 = np.arange(start=1, stop=13)
arr_2 = arr_2.reshape(4,3)
np.matmul(arr_1, arr_2)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr_np = np.random.random(49)
arr_np = arr_np.reshape(7,7)
arr_np
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=df[df['sex'] == 'Male'], x='total_bill', y='tip', hue='smoker')
plt.title('Male')
plt.subplot(1, 2, 2)
sns.scatterplot(data=df[df['sex'] == 'Female'], x='total_bill', y='tip', hue='smoker')
plt.title('Female')
plt.show()
```
